# Visualisation -IDVs(train data)

```python
def plot_feature(df, col_name, isContinuous):
  
    f, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12,3), dpi=90)
    
    # Plot without loan status
    if isContinuous:
        sns.distplot(df.loc[df[col_name].notnull(), col_name], color = 'orange', kde=False, ax=ax1)
    else:
        sns.countplot(df[col_name], order=sorted(df[col_name].unique()), color='#5975A4', saturation=1, ax=ax1)
    ax1.set_xlabel(col_name)
    ax1.set_ylabel('Count')
    ax1.set_title(col_name)
    plt.xticks(rotation = 90)

    # Plot with loan status
    if isContinuous:
        sns.boxplot(y=col_name, x='loan_default', data=df, ax=ax2)
        ax2.set_ylabel('')
        ax2.set_title(col_name + ' by  loan_default')
    else:
        data = df.groupby(col_name)['loan_default'].value_counts(normalize=True).to_frame('proportion').reset_index()        
        sns.barplot(x = col_name, y = 'proportion', hue= "loan_default", data = data, saturation=1, ax=ax2)
        ax2.set_ylabel('Loan ')
        ax2.set_title('Loan_default')
        plt.xticks(rotation = 90)
    #ax2.set_xlabel(col_name)
    
    plt.tight_layout()
train1 = pd.concat([train.reset_index(drop='True'),y_train.reset_index(drop='True')],axis=1)
```
